```python
import pandas as pd

#file names
pos_file = "../../raw/cedar_positive.csv"
neg_file = "../../raw/cedar_negative.csv"
output_file = "../../processed/cedar_combined.csv"
```

```python
#load files
def load_cedar(filepath, label): 
    with open(filepath, 'r') as f:
        first_line = f.readline() 
    sep = "\t" if "\t" in first_line else ","

    #read file
    df = pd.read_csv(filepath, sep=sep, dtype=str)

    #add label column
    df['label'] = label
    return df
```

```python
df_pos = load_cedar(pos_file, "Positive") 
df_neg = load_cedar(neg_file, "Negative")

#combine dataframes
df = pd.concat([df_pos, df_neg], ignore_index=True)
df.head() 
```

```python
new_cols = []

for col in df.columns: 
    col = col.replace(" - ", "_")
    col = col.replace(" ", "_")
    col = col.lower() 
    new_cols.append(col)

df.columns = new_cols
print(df.columns)
```

```python
df.head()
```

```python
#pick relevant columns 
clean = pd.DataFrame()

clean["mt_seq"] = df["epitope_name"].values
clean["wt_seq"] = df["related_object_name"].values
clean["mutation"] = df["epitope_mutation"].values
clean["epitope_relation"] = df["related_object_epitope_relation"].values
clean["source_molecule"] = df["epitope_source_molecule"].values
clean["label"] = df["label"].values

clean.head()
```

```python
#clean and filter

#checker
print(len(clean))

#uppercase and strip 
clean["mt_seq"] = clean["mt_seq"].str.strip().str.upper()
clean["wt_seq"] = clean["wt_seq"].str.strip().str.upper()

#drop rows with no peptide seq
clean = clean.dropna(subset=["mt_seq"]) 
print("after removing empty rows:", len(clean))
```

```python
#filter out invalid aa
valid_aa = set("ACDEFGHIKLMNPQRSTVWY")
def is_valid(seq): 
    if not isinstance(seq, str): 
        return False
    for aa in seq:
        if aa not in valid_aa: 
            return False
    return True
mask = clean["mt_seq"].apply(is_valid)
clean = clean[mask]
mask2 = clean["wt_seq"].apply(is_valid)
clean = clean[mask2]
print("after filtering invalid aa:", len(clean))
```

```python
#keep only inframe rows
mask = clean["epitope_relation"].str.contains("in-frame", case=False, na=True)
clean = clean[mask]
print("after filtering in-frame:", len(clean))
```

```python
#check if wildtype is valid
not_empty = clean["wt_seq"].notna()
valid_seq = clean["wt_seq"].apply(is_valid)

#combine
clean["has_wildtype"] = not_empty & valid_seq
print("after filtering wildtype:", len(clean))
```

```python
##REMOVE DUPLICATIONS

#positives per peptide
pos_count = {}

for i, row in clean.iterrows(): 
    pep = row["mt_seq"]
    label = row["label"]
    if pep not in pos_count: 
        pos_count[pep] = 0
    if label == "Positive":
        pos_count[pep] += 1

#drop duplicates
clean = clean.drop_duplicates(["mt_seq", "wt_seq"].copy())

#assign labels
new_labels = []
for pep in clean["mt_seq"]:
    if pos_count[pep] > 0: 
        new_labels.append("1")
    else: 
        new_labels.append("0")
clean["label"] = new_labels
print(len(clean))
```

```python
##SUMMARY
n_pos = 0
n_neg = 0
for label in clean["label"]:
    if label == "1":
        n_pos += 1
    else: 
        n_neg += 1
print("total peptides:", len(clean))
print("positive:", n_pos)
print("negative:", n_neg)
```
