```python
#load data
tcell_file = "../../raw/tcell_full_v3.csv"
mhc_file = "../../raw/mhc_ligand_full.csv"
```

```python
##load tcell and mhc exports
import pandas as pd 

#have 2 row headers
tcell_df = pd.read_csv(tcell_file, header=[0,1], low_memory = False)
tcell_df.columns = [f"{a} - {b}" for a,b in tcell_df.columns]
tcell_df.head()
```

```python
tcell_df.columns.tolist()
```

```python
mhc_df = pd.read_csv(mhc_file, header=[0,1], low_memory = True)
mhc_df.columns = [f"{a} - {b}" for a,b in mhc_df.columns]
mhc_df.head()
```

```python
mhc_df.columns.tolist()
```

```python
#mhc to keep

mhc_keep = [
    #IDENTIFIERS
    "Assay ID - CEDAR IRI", #trace rows
    "Reference - PMID", #reference 

    #EPITOPE INFO
    "Epitope - Epitope IRI", #merge dataset
    "Epitope - Name", #mutant peptide seq
    "Epitope - Mutation", 
    "Epitope - Source Molecule", #source protein
    "Epitope - Starting Position", 
    "Epitope - Ending Position", 

    #related(wt peptide)
    "Related Object - Epitope Relation", #to filter to frame shift
    "Related Object - Name", #wt seq
    "Related Object - Starting Position",
    "Related Object - Ending Position",

    #ASSAY INFO
    "Assay - Method", #different types = different results
    "Assay - Response measured",  
    "Assay - Qualitative Measurement", #positive/neg result
    "Assay - Quantitative measurement", #binding affinity

    #HLA INFO
    "MHC Restriction - Name", #hla allele
    "MHC Restriction - Class", 
    "MHC Restriction - Evidence Code"
    
]
```

```python
#keep only important columns

mhc_intermediate = mhc_df[mhc_keep].copy()
mhc_intermediate.head()
```

```python
#create clean epitope id
mhc_intermediate["epitope_id"] = mhc_intermediate["Epitope - Epitope IRI"].astype(str).str.split("/").str[-1]
mhc_intermediate.head()
```

```python
#save intermediate mhc 
mhc_intermediate.to_csv("../../processed/mhc_intermediate.csv", index=False)
```

```python
#check that rows are unaffected and columns change 

#checking rows and columns
print(len(mhc_df))
print(len(mhc_df.columns))

print(len(mhc_intermediate))
print(len(mhc_intermediate.columns))
```
