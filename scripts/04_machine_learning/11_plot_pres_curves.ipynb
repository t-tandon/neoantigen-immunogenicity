```python
import pickle
import csv

#load files
pickle_files = {
    "ESM2_LASSO": "../../output/01_lasso_pres/roc_esm2_pres.pkl", 
    "ESM2_RF": "../../output/01_rf_pres/roc_esm2_rf_pres.pkl", 
    "Prot_LASSO": "../../output/01_lasso_pres/roc_prot_pres.pkl", 
    "Prot_RF": "../../output/01_rf_pres/roc_prot_rf_pres.pkl", 
}
```

```python
#load pickle, one row per curve plot

with open("../../output/roc_curves_pres.csv", "w", newline="") as csv_file: 
    writer = csv.writer(csv_file)

    #column names
    writer.writerow([
        "source", "dataset", "feature_set", "fpr", "tpr", "tpr_std"
    ])
    row_count = 0 

    #go through each pickle file
    for source, filename in pickle_files.items(): 
        #load data
        with open(filename, "rb") as file: 
            curves = pickle.load(file)
        print("loaded", source)
        print("curves", len(curves))

        #each curve is dict
        for curve in curves: 
            #one row for every plot
            for fpr, tpr, std in zip(
                curve["mean_fpr"], 
                curve["mean_tpr"], 
                curve["std_tpr"]
            ): 
                writer.writerow([
                    source, curve["dataset"], 
                    curve["feature_set"], fpr, tpr, std
                ])
                row_count +=1 
print("wrote", row_count)
```
