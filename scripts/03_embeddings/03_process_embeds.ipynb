```python
import numpy as np 

#paths
input_path = "../../output/embeddings/esm2_embeddings.npz"
output_path = "../../output/final_embeds/esm2_final.npz"

#load data
data = np.load(input_path, allow_pickle=True)
#load features 
n = int(data["n_peptides"])
labels = data["labels"]
lengths = data["lengths"]
epitope_ids = data["epitope_ids"]
mt_eos = data["mt_eos"]
wt_eos = data["wt_eos"]

#since esm2 has bos also 
mt_bos = data["mt_bos"]
wt_bos = data["wt_bos"]

#also load per residue 
mt_perres = [data[f"mt_perres_{i}"] for i in range(n)]
wt_perres = [data[f"wt_perres_{i}"] for i in range(n)]
```

```python
#filter for same length peptides
keep_idx = [i for i in range(n) if mt_perres[i].shape[0] == wt_perres[i].shape[0]]
print(f"Keeping {len(keep_idx)} out of {n} peptides with same length.")

#apply filter
labels = labels[keep_idx]
lengths = lengths[keep_idx]
epitope_ids = [epitope_ids[i] for i in keep_idx]
mt_eos = mt_eos[keep_idx]
wt_eos = wt_eos[keep_idx]
mt_perres = [mt_perres[i] for i in keep_idx]
wt_perres = [wt_perres[i] for i in keep_idx]
mt_bos = mt_bos[keep_idx]
wt_bos = wt_bos[keep_idx]

n_kept = len(keep_idx)

#check
print("NaNS mt_eos:", np.isnan(mt_eos).sum())
print("NaNS wt_eos:", np.isnan(wt_eos).sum())
print("NaNS mt_bos:", np.isnan(mt_bos).sum())
print("NaNS wt_bos:", np.isnan(wt_bos).sum())
```

```python
#compute differences
diff_eos = (mt_eos - wt_eos).astype(np.float32)
diff_bos = (mt_bos - wt_bos).astype(np.float32)
```

```python
#save dictionary 
save_dict = {
    "labels": labels,
    "lengths": lengths,
    "n_peptides": np.array(n_kept), 
    "kept_indices": np.array(keep_idx),
    "epitope_ids": np.array(epitope_ids),
    "mt_eos": mt_eos, 
    "wt_eos": wt_eos,
    "diff_eos": diff_eos,
    #same for bos
    "mt_bos": mt_bos,
    "wt_bos": wt_bos,
    "diff_bos": diff_bos
}

#for per residue, save as list of arrays
for i in range(n_kept):
    save_dict[f"mt_perres_{i}"] = mt_perres[i]
    save_dict[f"wt_perres_{i}"] = wt_perres[i]
    save_dict[f"diff_perres_{i}"] = (mt_perres[i] - wt_perres[i]).astype(np.float32)

#save
np.savez_compressed(output_path, **save_dict)
```
